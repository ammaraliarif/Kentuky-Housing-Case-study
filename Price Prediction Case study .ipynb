```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
linreg = LinearRegression()
```

```python
tdata = data.drop(['id','price','date'],axis=1)
conv_dates = [1 if values == 2014 else 0 for values in data.date]
labels = data['price']
#tdata['date'] = conv_dates
```

```python
X_train , X_test , y_train , y_test = train_test_split(tdata , labels, test_size = 0.2 , random_state = 42 )
```

```python
linreg.fit(X_train,y_train)
```

```python
linreg.score(X_test,y_test)
```

```python
#Cross Validation 
cv_results = cross_val_score(linreg , tdata, labels, cv=5)
print(cv_results)
print(np.mean(cv_results))
```

```python
# Lasso Feature selection
#removing the alpha generates similar score otherwise alpha=0.1
#since no covergence warning is given when alpha is remved I removed the alpha and let sklearn figure it out 
lasso = Lasso(normalize=True)
names = tdata.columns
lasso_coef = lasso.fit(tdata, labels).coef_
plt.plot(range(len(names)), lasso_coef)
plt.xticks(range(len(names)), names,rotation = 90)
plt.ylabel('Coefficients')
plt.show()
```

```python
lasso.fit(X_train,y_train)
lasso_pred = lasso.predict(X_test)
lasso_score= lasso.score(X_test,y_test)
cv_results_lasso = cross_val_score(lasso, tdata, labels, cv=5)
print(lasso_score)
print(cv_results_lasso)
print(np.mean(cv_results_lasso))
```

```python
#Ridge regression
ridge = Ridge(alpha=0.1,normalize=True)
ridge.fit(X_train , y_train)
ridge_pred = ridge.predict(X_test)
ridge_score=ridge.score(X_test, y_test)
cv_results_ridge= cross_val_score(ridge, tdata, labels, cv=5)
print(ridge_score)
print(cv_results_ridge)
print(np.mean(cv_results_ridge))
```
